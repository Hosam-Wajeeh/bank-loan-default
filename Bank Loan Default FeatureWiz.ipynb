# Insights from the CatPlots:

*Loans that had a 36 months term were slightly better at being Fully Paid(0.12) than loans with 60 months term(0.24). Also the number of loans that were taken on a 36 month term in the dataset were double the loans that were taken on a 60 month term (Check the countplot Below).

*Better grades of a loan had a lower chance at being Charged Off.

*Subgrade had the same insights as grade BUT note that in grade G had the highest chance of being charged off of all the grades and in sub_grade F5 and F4 had a higher chance of being charged off than G5 and G4 while F3,F2 and F1 had a lower chance at being charged off than G3,G2 and G1.

*Home ownership were all close to each other while OTHER and NONE categories slightly higher at being Charged Off. Also the lowest number of loans handed had OTHER and NONE categories as their home_ownership (Check the countplot Below).

*Verification status: All close to each other.

*Credit Requirements: Yes had more chance at being Fully Paid than No. (Check the number of loans that had NO as credit_requirement from the countplot below).

*Purpose: small business category had the highest chance of being Charged Off while the cars category had the highest probability of being Fully Paid.

*addr_state: Loans with NE state had a very high chance to be Charged Off while loans with ME as state had the lowest chance of being Charged Off.

```python
threshold = 0.80    #correlation threshhold
correlated_features = set()
correlation_matrix = X_train.corr()
for i in range(len(correlation_matrix .columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[j, i]) >= threshold:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)
            
len(correlated_features)
```

# Model 1

### Based on all the Features - XGBoost

```python
#Fitting a baseline model to figure out feature importances
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
def print_results(results):
    print('BEST PARAMS: {}\n'.format(results.best_params_))

    means = results.cv_results_['mean_test_score']
    for mean, params in zip(means, results.cv_results_['params']):
        print('{} for {}'.format(round(mean, 3), params))
```

```python
X_train_clean = selected_features_df.drop('loan_status_labeled', axis=1)
X_train_clean.head(1)
```

```python
parameters = {
    'n_estimators': [2**i for i in range(3, 10)],
    'max_depth': [2, 4, 8, 16, 32, None]
}

from xgboost import XGBClassifier
xgb = XGBClassifier(use_label_encoder=False, random_state=0)
model = GridSearchCV(xgb, parameters, cv=5, n_jobs=-1)
model.fit(X_train_clean, y_train.values.ravel())

print_results(model)
```

```python
# Generate feature importance plot
plt.figure(figsize=(15,10))
feat_imp = model.best_estimator_.feature_importances_
indices = np.argsort(feat_imp)
plt.yticks(range(len(indices)), [X_train_clean.columns[i] for i in indices])
plt.barh(range(len(indices)), feat_imp[indices], color='r', align='center')
plt.show()
```
